# train/test set   

```python

df_match = pd.read_csv('match_drug_arr_exclusion_20201110.csv',sep=',')

set_value = []        
tests=0
trains=0
drops=0
for y in range(len(df_match)):
    if str(y) in test_row:
        set_value.append('1')
        tests +=1
    elif str(y) in train_row:
        set_value.append('0')
        trains +=1        
    else:
        set_value.append('2')
        drops +=1

df_match['set_value']=set_value
print('tests:',tests)
print('trains:',trains)
print('drops:',drops)

df_match.to_csv('match_setvalue_20201201.csv',sep=',')
print(df_match)
```

```python
df_match.replace('2',np.nan,inplace=True)
df_match.dropna(subset=['set_value'],inplace=True)
df_match=df_match.reset_index()
print(df_match)
```

```python
opid = df_match['opid'].tolist()

single = []
dupl=[]
for d in range(len(opid)):
    if opid[d] in single:
        dupl.append(d)
        continue
    else:
        single.append(opid[d])
print(dupl)
```

```python
mang=[]
mangidx=[]
for dd in dupl:
    wth = df_match[df_match['opid']==df_match.loc[dd,'opid']].index.tolist()
    wth.remove(dd)
    for wtf in wth:
        if df_match.loc[dd,'path'] != df_match.loc[wtf,'path']:
            mang.append(df_match.loc[dd,'opid'])
            mangidx.append(dd)

#print(mang)
manga = np.array(mang)
manga = np.unique(manga)
print(manga.size)
```

```python
idmiss = df_match.loc[mangidx,['opid','path']]
idmiss.to_csv('idmiss.csv')
```
