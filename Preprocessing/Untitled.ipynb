```python
import pandas as pd
import pickle

df_preop = pickle.load(open('df_preop_match','rb'))
old = pickle.load(open('../DL_model/caseid_test','rb'))
new = pickle.load(open('../DL_model/caseid_test_new','rb'))
```

```python
for oid in old:
    if oid not in new:
        print(oid)
```

```python
df_preop

for i in range(len(df_preop)):
    df_preop.loc[i,'diff']=0
    if df_preop.loc[i,'file_path'] in old:
        df_preop.loc[i,'old']=1
    if df_preop.loc[i,'file_path'] in new:
        df_preop.loc[i,'new']=1
    if df_preop.loc[i,'old']==1 or df_preop.loc[i,'new']==1:
        if df_preop.loc[i,'old'] != df_preop.loc[i,'new']:
            df_preop.loc[i,'diff']=1
```
