```python
# preop data(NRS 0)  spi 
file_spi_test_preop = pickle.load(open('../DL_model/file_spi_test_preop','rb'))
temps = np.array(pd.Series(file_spi_test_preop).str.split(',', n=1).tolist())[:,1]
vital_test_preop = [temp[:-3]+'vital' for temp in temps]

spi_y_pred_preop = []
# preop test data  spi 
for idx, pleth in enumerate(x_test_preop):
    print('{}/{}...'.format(idx+1, len(x_test_preop)), end='')
    
    f_vital = vital_test_preop[idx]
    
    
    # individual histogram
    hist_path = 'cache/histogram_preop/'+f_vital
    if os.path.exists(hist_path):
        hist_hbi, hist_ppga = pickle.load(open(hist_path,'rb'))
    
    #   SPI  2    
    else:
        spi_y_pred_preop.append(-1)
        continue
    
    spi_per_input = []
        
    #  HBI, PPGA      
    print('detecting peaks...',end='')
    minlist, maxlist = arr.detect_peaks(pleth, 100)

    hbi_list, ppga_list = [], []
    for j in range(len(maxlist) - 1):
        # HBI, PPGA   
        hbi = (maxlist[j+1] - maxlist[j]) / srate * 1000
        ppga = pleth[maxlist[j+1]] - pleth[minlist[j]]
        if 240 < hbi < 2000:
            hist_hbi.learn(hbi)
            hbi_list.append(hbi)
        else:
            if 0 < ppga < 100:
                hist_ppga.learn(ppga)
                continue
            continue

        if 0 < ppga < 100:
            hist_ppga.learn(ppga)
            ppga_list.append(ppga)    
        else:
            continue

        # HBI PPGA norm
        hbi_perc = hist_hbi.percentile(hbi)*0.73 + group_hist_hbi_preop.percentile(hbi)*0.27
        ppga_perc = hist_ppga.percentile(ppga)*0.73 + group_hist_ppga_preop.percentile(ppga)*0.27

        # SPI 
        spi = 100 - (0.7*ppga_perc + 0.3*hbi_perc)
        spi_per_input.append(spi)

    # input   hbi, ppga  
    if spi_per_input == []:
        spi_y_pred.append(-2)
        continue

    spi_y_pred_preop.append(np.mean(spi_per_input)/10)
    print('done')
```

```python
spi_y = np.concatenate((spi_y_pred, spi_y_pred_preop))
false_mask = spi_y < 0

y_test = np.load('../DL_model/dataset/y_test_comp.npz')['arr_0']
print('Regression of SPI model:', np.mean(np.around(spi_y[~false_mask]) == np.around(y_test[~false_mask])))

print('severe pain prediction of SPI model:', np.mean((spi_y[~false_mask]>=5) == (y_test[~false_mask]>=5)))
```
