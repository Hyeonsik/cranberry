## basic model

```python
from keras.models import Sequential
from keras.models import Model, load_model
from keras.optimizers import Adam as Adam
from keras.layers import Dense, Conv1D, MaxPooling1D, GlobalMaxPool1D, BatchNormalization, Dropout, Activation
from keras.layers import GlobalAveragePooling1D, Flatten, SeparableConv1D, LeakyReLU, Add
from keras.layers import Input, concatenate
from keras.callbacks import ModelCheckpoint, EarlyStopping
from sklearn.metrics import auc, classification_report, confusion_matrix, accuracy_score, roc_curve, roc_auc_score, f1_score, precision_recall_curve
import tensorflow as tf
import os, pickle
from keras import backend as K
from keras import regularizers
from keras.initializers import he_normal, GlorotNormal
import tensorflow_addons as tfa
from keras_radam import RAdam
from keras_radam.training import RAdamOptimizer

# hyperparamters
num_nodes = [32,64,128] #, 64, 64, 64]
kernel_size = 3
kernels = [3,3,3]
strides=[1,1,1]
pool_size = 2
BATCH_SIZE = 64
dense_node = 64
dense_node2 = 0
dropout_cnn = 0.4
dropout_fc = 0.2
dropout_fc2 = 0
learning_rate = 0.001


# initializer & regularizer
k_init = he_normal(seed=None)
#k_init = GlorotNormal(seed=None) # Xavier normal initializer
#k_init = None
b_init = None
k_l2 = 0
#k_reg = regularizers.l2(l2=k_l2)
k_reg = None
b_reg = None


testname = '-'.join([str(num_node) for num_node in num_nodes])
print(testname)


# 출력 폴더를 생성
model_name = '1D_CNN_PPG50Hz_bin_train_w_samp3_agender_RAdam'
for num_node in num_nodes:
    model_name += 'conv{}x2_'.format(num_node)
for kernel in kernels:
    model_name += 'filt{}_'.format(kernel)

model_name += 'bn_maxpool{}_globalaverage_pool_dropout{}_dense{}_dropout{}_dense{}_dropout{}_batch{}_lr_{}_kinit'.format(strides[0], dropout_cnn, dense_node, dropout_fc, dense_node2, dropout_fc2, BATCH_SIZE, learning_rate)

#model_name = 'model_reg_{}_{}_{}_size{}_relu_bn_maxpool{}_globalmaxpool_dense32_dropout{}_batch{}_learning_rate{}'.format(num_nodes[0], num_nodes[1], num_nodes[2], kernel_size, pool_size, dropout_rate, BATCH_SIZE, learning_rate)
save_path = "output/preprocess6/input/"+model_name

if not os.path.exists(save_path):
    os.mkdir(save_path)
weight_path = save_path + "/weights.hdf5"


# GPU 설정
strategy = tf.distribute.MirroredStrategy(devices=["/gpu:0","/gpu:1"])
with strategy.scope():
    n_act=0
    
    
    # build a model
    inp_fnn = Input(shape=(x_trains[1].shape[1],))
    inp_cnn = Input(shape=(x_trains[0].shape[1], x_trains[0].shape[2]))
    
    out_fnn = inp_fnn
    out_fnn = Activation('sigmoid', name='activation{}'.format(n_act)) (out_fnn)
    n_act+=1
    
    out_cnn = inp_cnn
    for idx, num_node in enumerate(num_nodes):
        out_cnn = Conv1D(filters=num_node, kernel_size=kernels[idx], strides=strides[idx], padding='same', kernel_initializer=k_init) (out_cnn)
        out_cnn = Conv1D(filters=num_node, kernel_size=kernels[idx], padding='same', kernel_initializer=k_init) (out_cnn)
        out_cnn = Activation('relu', name='activation{}'.format(n_act)) (out_cnn)  # out_cnn = LeakyReLU(alpha=0.1) (out_cnn)
        n_act+=1
        out_cnn = BatchNormalization() (out_cnn)
        out_cnn = MaxPooling1D(pool_size=pool_size) (out_cnn)
        
        
    #out_cnn = GlobalMaxPool1D() (out_cnn)
    out_cnn = GlobalAveragePooling1D() (out_cnn)
    out_cnn = Dropout(dropout_cnn) (out_cnn)
    
    out = concatenate([out_fnn, out_cnn])
    
    if dense_node != 0:
        out = Dense(dense_node, kernel_initializer=k_init, activation='tanh', name='activation{}'.format(n_act)) (out) # age값이 너무 확튀면 sigmoid?
        n_act+=1
        out = Dropout(dropout_fc) (out)
        
    if dense_node2 !=0:
        out = Dense(dense_node2, kernel_initializer=k_init, activation='tanh', name='activation{}'.format(n_act)) (out) # age값이 너무 확튀면 sigmoid?
        n_act+=1
        out = Dropout(dropout_fc2) (out)
        
    out = Dense(1, activation='sigmoid', name='activation{}'.format(n_act)) (out)
    n_act+=1
    
    model = Model(inputs=[inp_cnn, inp_fnn], outputs=[out])
    
    print(model.summary())
    
    # model 학습 설정
    #model.compile(RAdam(name='RectifiedAdam'), loss='binary_crossentropy', metrics=[tf.keras.metrics.AUC()])
    model.compile(loss='binary_crossentropy', optimizer=Adam(learning_rate=learning_rate), metrics=["acc", tf.keras.metrics.AUC()])
    hist = model.fit(x_trains, y_train_bin, validation_data = (x_vals, y_val_bin, val_w_samp), sample_weight=train_w_samp, epochs=100, batch_size=BATCH_SIZE, #class_weight={0:1, 1:3}, 
                            callbacks=[ModelCheckpoint(monitor='val_loss', filepath=weight_path, verbose=1, save_best_only=True),
                                        EarlyStopping(monitor='val_loss', patience=5, verbose=0, mode='auto')])

    
    #tf.keras.backend.clear_session()
    model.load_weights(weight_path)

    # 모델의 아키텍처 및 구조 저장
    open(save_path + "/model.json", "wt").write(model.to_json())


    # 전체 test 샘플을 한번에 예측
    y_pred = model.predict(x_tests).flatten() # binary는 flatten

# 결과를 저장
#np.savetxt(save_path+'/pred_y.txt', y_pred)


# 모델의 history log 저장 - binary classification
for key in hist.history.keys():
    if 'auc' in key and not 'val' in key:
        auc_key = key
#pickle.dump((hist.history['loss'], hist.history['val_loss'], hist.history['accuracy'], hist.history['val_accuracy'], hist.history[auc], hist.history['val_'+auc]), open(save_path+'/history', 'wb'))

# 모델의 history log 저장 - regression
pickle.dump((hist.history['loss'], hist.history['val_loss'], hist.history['acc'], hist.history['val_acc'], hist.history[auc_key], hist.history['val_'+auc_key]), open(save_path+'/history', 'wb'))
```

```python
from keras.models import Sequential
from keras.models import Model, load_model
from keras.optimizers import Adam as Adam
from keras.layers import Dense, Conv1D, MaxPooling1D, GlobalMaxPool1D, BatchNormalization, Dropout, Activation
from keras.layers import GlobalAveragePooling1D, Flatten, SeparableConv1D, AveragePooling1D, Input, concatenate, add, LSTM, Average
from keras.callbacks import ModelCheckpoint, EarlyStopping
from sklearn.metrics import auc, classification_report, confusion_matrix, accuracy_score, roc_curve, roc_auc_score, f1_score, precision_recall_curve
import tensorflow as tf
import os, pickle
from keras import backend as K


def inception(inp, nfilters=32):
    # 각각은 길이가 줄어들지 않는다

    # 1 conv
    c1 = Conv1D(filters=nfilters, kernel_size=1, kernel_initializer='glorot_uniform', padding='same', activation='relu')(inp)

    # 3 reduce
    c2 = Conv1D(filters=nfilters, kernel_size=1, kernel_initializer='glorot_uniform', padding='same', activation='relu')(inp)
    # 3 conv
    c2 = Conv1D(filters=nfilters, kernel_size=3, kernel_initializer='glorot_uniform', padding='same', activation='relu')(c2)

    # 5 reduce
    c3 = Conv1D(filters=nfilters, kernel_size=1, kernel_initializer='glorot_uniform', padding='same', activation='relu')(inp)
    # 5 conv
    c3 = Conv1D(filters=nfilters, kernel_size=5, kernel_initializer='glorot_uniform', padding='same', activation='relu')(c3)

    # pooling
    m1 = MaxPooling1D(pool_size=3, strides=1, padding='same')(inp)
    # pooling proj
    m1 = Conv1D(filters=nfilters, kernel_size=1, kernel_initializer='glorot_uniform', padding='same', activation='relu')(m1)

    return concatenate([c1, c2, c3, m1], axis=-1)


def inception_ext(inp, nfilters=32):
    # 각각은 길이가 줄어들지 않는다

    # 1 conv
    c1 = Conv1D(filters=nfilters, kernel_size=1, kernel_initializer='glorot_uniform', padding='same', activation='relu')(inp)

    # 3 reduce
    c2 = Conv1D(filters=nfilters, kernel_size=1, kernel_initializer='glorot_uniform', padding='same', activation='relu')(inp)
    # 3 conv
    c2 = Conv1D(filters=nfilters, kernel_size=3,  dilation_rate = 5, kernel_initializer='glorot_uniform', padding='same', activation='relu')(c2)

    # 5 reduce
    c3 = Conv1D(filters=nfilters, kernel_size=1, kernel_initializer='glorot_uniform', padding='same', activation='relu')(inp)
    # 5 conv
    c3 = Conv1D(filters=nfilters, kernel_size=5,  dilation_rate = 7, kernel_initializer='glorot_uniform', padding='same', activation='relu')(c3)

    # pooling
    m1 = MaxPooling1D(pool_size=3, strides=1, padding='same')(inp)
    # pooling proj
    m1 = Conv1D(filters=nfilters, kernel_size=1, kernel_initializer='glorot_uniform', padding='same', activation='relu')(m1)

    return concatenate([c1, c2, c3, m1], axis=-1)
```

```python
from keras import backend as K


# squeeze-excitation residual block (SE-Resnet module)
def SE_block(inp, r):
    
    # residual path
    inp0 = inp
    
    
    # squeeze
    inp = GlobalAveragePooling1D() (inp)
    
    # compress
    size = K.int_shape(inp)[1]
    inp = Dense(units=size//r, activation='relu') (inp)
    
    # decompress
    inp = Dense(units=size, activation='sigmoid') (inp)
    
    return Multiply() ([inp0, inp])


# non-local module
def non_local_module(inp):
    
    # direct path
    inp0 = inp
    
    c1 = Conv1D(filters=128, kernel_size=1, strides=1, padding='same', kernel_initializer=k_init, activation='relu') (inp)
    
    #c2 = Dot(axes=1) ([c1, c1])
    c2 = Multiply() ([c1, c1])
    c2 = Activation('softmax') (c2)
    
    #c3 = Dot(axes=1) ([c2, c1])
    c3 = Multiply() ([c2, c1])
    c3 = Conv1D(filters=128, kernel_size=1, strides=1, padding='same', kernel_initializer=k_init, activation='relu') (c1)
    
    return Add() ([inp0, c3])
```

```python
def extraction_block(inp, kernel_size=7, nfilters=8):
    c1 = Conv1D(filters=nfilters, kernel_size=kernel_size, dilation_rate = 1, kernel_initializer='glorot_uniform', padding='same', activation='relu')(inp)
    c2 = Conv1D(filters=nfilters, kernel_size=kernel_size, dilation_rate = 2, kernel_initializer='glorot_uniform', padding='same', activation='relu')(inp)
    c3 = Conv1D(filters=nfilters, kernel_size=kernel_size, dilation_rate = 3, kernel_initializer='glorot_uniform', padding='same', activation='relu')(inp)
    c4 = Conv1D(filters=nfilters, kernel_size=kernel_size, dilation_rate = 4, kernel_initializer='glorot_uniform', padding='same', activation='relu')(inp)
    
    c5 = concatenate([c1, c2, c3, c4], axis=-1)
    c5 = Conv1D(filters=nfilters, kernel_size=1, kernel_initializer='glorot_uniform', padding='same', activation='relu')(c5)
    
    c6 = Add() ([inp, c5])
    c6 = BatchNormalization() (c6)
    
    return Activation('relu') (c6)


def concentration_block(inp, kernel_size=11, nfilters=32):
    out = Conv1D(filters=nfilters, kernel_size=kernel_size, strides=2, padding='same', kernel_initializer=k_init, activation='relu') (inp)
    out = BatchNormalization() (out)
    
    return Activation('relu') (out)
```

```python
from keras.models import Sequential
from keras.models import Model, load_model
from keras.optimizers import Adam as Adam
from keras.layers import Dense, Conv1D, MaxPooling1D, GlobalMaxPool1D, BatchNormalization, Dropout, Activation
from keras.layers import GlobalAveragePooling1D, Flatten, SeparableConv1D, LeakyReLU
from keras.layers import Input, concatenate, Add, Multiply, AveragePooling1D, Dot
from keras.callbacks import ModelCheckpoint, EarlyStopping, LearningRateScheduler
from sklearn.metrics import auc, classification_report, confusion_matrix, accuracy_score, roc_curve, roc_auc_score, f1_score, precision_recall_curve
import tensorflow as tf
import os, pickle
from keras import backend as K
from keras import regularizers
from keras.initializers import he_normal, GlorotNormal
import tensorflow_addons as tfa

# hyperparamters
kernel_size = 3
pool_size = 2
BATCH_SIZE = 1024
dense_node = 64
dense_node2 = 0
dropout_cnn = 0.4
dropout_fc = 0.2
dropout_fc2 = 0
dropout_rate = 0.2


# learning rate
learning_rate = 0.001

from keras.callbacks import LearningRateScheduler
def decay_schedule(epoch, lr):
    # decay by 0.5 every 5 epochs; use `% 1` to decay after each epoch
    if (epoch % 5 == 0) and (epoch != 0):
        lr = lr * 0.5
    return lr

lr_scheduler = LearningRateScheduler(decay_schedule)


# initializer & regularizer
k_init = he_normal(seed=None)
#k_init = GlorotNormal(seed=None) # Xavier normal initializer
#k_init = None
b_init = None
k_l2 = 0
#k_reg = regularizers.l2(l2=k_l2)
k_reg = None
b_reg = None



# 출력 폴더를 생성
model_name = '1D_CNN_PPG25Hz_bin_train_w_samp3_S_Baek_'
save_path = "output/preprocess6/input/"+model_name
print(model_name)


if not os.path.exists(save_path):
    os.mkdir(save_path)
weight_path = save_path + "/weights.hdf5"


# GPU 설정
strategy = tf.distribute.MirroredStrategy(devices=["/gpu:0","/gpu:1"])
with strategy.scope():
    
    # build a model
    inp_fnn = Input(shape=(x_trains[1].shape[1],))
    inp_cnn = Input(shape=(x_trains[0].shape[1], x_trains[0].shape[2]))
    
    out_fnn = inp_fnn
    out_fnn = Activation('sigmoid') (out_fnn)
    
    
    out_cnn = inp_cnn
    
    
    # first block
    out_cnn = extraction_block(out_cnn, kernel_size=7, nfilters=8)
    out_cnn = Dropout(dropout_rate) (out_cnn)
    out_cnn = concentration_block(out_cnn, kernel_size=11, nfilters=32)
    out_cnn = Dropout(dropout_rate) (out_cnn)

    # second block
    out_cnn = extraction_block(out_cnn, kernel_size=7, nfilters=32)
    out_cnn = Dropout(dropout_rate) (out_cnn)
    out_cnn = concentration_block(out_cnn, kernel_size=11, nfilters=64)
    out_cnn = Dropout(dropout_rate) (out_cnn)


    # third block
    out_cnn = extraction_block(out_cnn, kernel_size=7, nfilters=64)
    out_cnn = Dropout(dropout_rate) (out_cnn)
    out_cnn = concentration_block(out_cnn, kernel_size=11, nfilters=128)
    out_cnn = Dropout(dropout_rate) (out_cnn)

    # forth block
#    out_cnn = extraction_block(out_cnn, kernel_size=7, nfilters=128)
#    out_cnn = Dropout(dropout_rate) (out_cnn)
#    out_cnn = concentration_block(out_cnn, kernel_size=11, nfilters=256)
#    out_cnn = Dropout(dropout_rate) (out_cnn)    
     

    # FCN
    out_cnn = GlobalAveragePooling1D() (out_cnn)
    #out_cnn = Dropout(dropout_cnn) (out_cnn)
    
    out = concatenate([out_fnn, out_cnn])
    
    out = Dense(dense_node, kernel_initializer=k_init, activation='tanh') (out) # age값이 너무 확튀면 sigmoid?
    out = Dropout(dropout_fc) (out)
        
    if dense_node2 !=0:
        out = Dense(dense_node2, kernel_initializer=k_init, activation='tanh') (out) # age값이 너무 확튀면 sigmoid?
        out = Dropout(dropout_fc2) (out)
        
        
    out = Dense(1, activation='sigmoid') (out)
    
    
    model = Model(inputs=[inp_cnn, inp_fnn], outputs=[out])
    
    
    print(model.summary())
    
    # model 학습 설정
    model.compile(loss='binary_crossentropy', optimizer=Adam(lr=learning_rate), metrics=["acc", tf.keras.metrics.AUC()])
    hist = model.fit(x_trains, y_train_bin, validation_data = (x_vals, y_val_bin, val_w_samp), sample_weight=train_w_samp, epochs=100, batch_size=BATCH_SIZE, #class_weight={0:1, 1:3}, 
                            callbacks=[ModelCheckpoint(monitor='val_loss', filepath=weight_path, verbose=1, save_best_only=True),
                                        EarlyStopping(monitor='val_loss', patience=15, verbose=0, mode='auto'), lr_scheduler])


    # 모델의 아키텍처 및 구조 저장
    open(save_path + "/model.json", "wt").write(model.to_json())


    # best weight을 loading
    model.load_weights(weight_path)
    # 전체 test 샘플을 한번에 예측
    y_pred = model.predict(x_tests).flatten() # binary는 flatten

    
# 결과를 저장
#np.savetxt(save_path+'/pred_y.txt', y_pred)


# 모델의 history log 저장 - binary classification
for key in hist.history.keys():
    if 'auc' in key and not 'val' in key:
        auc_key = key
#pickle.dump((hist.history['loss'], hist.history['val_loss'], hist.history['accuracy'], hist.history['val_accuracy'], hist.history[auc], hist.history['val_'+auc]), open(save_path+'/history', 'wb'))

# 모델의 history log 저장 - regression
pickle.dump((hist.history['loss'], hist.history['val_loss'], hist.history['acc'], hist.history['val_acc'], hist.history[auc_key], hist.history['val_'+auc_key]), open(save_path+'/history', 'wb'))
```
